```python
# Image Processing
1. Colored Image 3: Red, Green,Bluw
2. Gray Scaling - 1 channel
3. Pixel range in between 0-255
```

```python
image_path="/content/Alphabet Example.jpg"
print(image_path)
```

```python
import cv2
img1=cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)
img1
```

```python
import matplotlib.pyplot as plt
plt.imshow(img1,cmap="gray")
```

```python
# Visualizing the pixel
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(25,25))
sns.heatmap(img1,cmap="gray",annot=True)
```

```python
image_path2="/content/CuteCat.jpeg"
img2=cv2.imread(image_path2,cv2.IMREAD_COLOR_RGB)

img2
```

```python
# Show in blue color
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```
